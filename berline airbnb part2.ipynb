## Let us consider the room type and calculate the reviews average in all the dimensions

###### first calculating how many types of rrom type are there and total number of data related to it

```python
print('\n from the analysis we can predict that the total number of each room type is as follows')
print(pd.value_counts(berlins.room_type))
a=pd.value_counts(berlins.room_type)
print('\n Total data\'s related to private room are:',a[0])
print('\n Total data\'s related to entire room or apartment are:',a[1])
print('\n Total data\'s related to shared room are:',a[2])
```

###### Now calculating the total reviews based on the room types

```python
dimension=[]
dimensions=['number_of_reviews','price','reviews_per_month','review_scores_value','review_scores_location','review_scores_communication','review_scores_checkin','review_scores_cleanliness','review_scores_rating','review_scores_accuracy']
for j in range(len(dimensions)):
    a=berlins[berlins['room_type']=='Private room'][dimensions[j]].mean()
    dimension.append(a)
    print('The average reviews of berlin by ',dimensions[j],'for Private room is : ',a)
print(sb.distplot(dimension))
```
